```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('Screenshot 2025-10-30 152834.png')
```

```python
plt.imshow(image[:,:,::-1])
plt.title('Original Image')
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
medium=np.median(image)
```

```python
low=int(max(0,0.7*medium))
high=int(max(255,1/3*medium))
         
```

```python
blurred_gray = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

```python
plt.subplot(1, 2, 2)
plt.imshow(blurred_gray, cmap='gray')
plt.title('Blurred')
plt.axis('off')

plt.show()
```

```python
edges = cv2.Canny(gray_image, 90, 140)
```

```python
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```
